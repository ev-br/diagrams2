## Энергия полярона для W=2, L = 60

```python
plt.figure(figsize=(12, 7))
plt.plot(range(1,120), E, 'o--', color='orange')
plt.xlabel('N')
plt.ylabel('E')
plt.show()
```

## 3/4

```python
E40_3_4 = (-61.38218273347809 -  -53.62903482654785 - -2.9941316023674793)
E60_3_4 = df1[(df1['Nup_total']==45) & (df1['Ndown_total']==1)]['E'].values[0] - \
    df1[(df1['Nup_total']==45) & (df1['Ndown_total']==0)]['E'].values[0] - \
    df1[(df1['Nup_total']==1) & (df1['Ndown_total']==0)]['E'].values[0]
E80_3_4 = -115.53714219677586 - -107.77080699996519 - -2.998495905008462
E120_3_4 = -169.66808988788966 - -161.903284122513 - -2.9993259306067452
```

```python
import lmfit
from lmfit.models import ExpressionModel

mod_e = ExpressionModel('a*x+b')

# Устанавливаем начальные значения параметров (с которых стартует оптимизация)
mod_e.set_param_hint("a", value=1.0, vary=True)
mod_e.set_param_hint("b", value=1.0, vary=True)

_1_L = [1/60,1/80, 1/120]
E = [E60_3_4, E80_3_4, E120_3_4]

params_e = mod_e.make_params()

result_e = mod_e.fit(E, params_e, method="leastsq", x=_1_L)  # fitting

print('result for e=',result_e.best_values['b'])
```

```python
plt.figure(figsize=(12, 7))
# plt.plot([h40_2,h60_2],[mu40_2,mu60_2], 'o--', color='black', label='N=2')
plt.plot([1/40, 1/60, 1/80, 1/120],
         [E40_3_4,E60_3_4, E80_3_4, E120_3_4],
         'o--', color='orange', label='N_3/4')
plt.plot([0,1/120],[result_e.best_values['b'],E120_3_4], 'o--',color='black')
plt.annotate(xy=[1/40,E40_3_4], s='40')
plt.annotate(xy=[1/60,E60_3_4], s='60')
plt.annotate(xy=[1/80,E80_3_4], s='80')
plt.annotate(xy=[1/120,E120_3_4], s='120')
plt.annotate(xy=[0+0.001,result_e.best_values['b']], s='Extrapolation'+ f" {result_e.best_values['b']}")
plt.xlabel('1/L')
plt.ylabel('E')
plt.show()
```
