```python
import xmltodict
import numpy as np
import matplotlib.pyplot as plt
```

```python
def En(j, n, m, folder='t-1/'):
    Energy = [] 
    path = "C:/Users/Mindelly/Documents/GitHub/diagrams2/w2/" + folder
    for i in range(n, m, 2):
        file_name = str(i) +'-'+ str(j) + ".task1.out.xml"
        with open(path + file_name) as xml_file:
            data_dict = xmltodict.parse(xml_file.read())
        Energy.append(float(data_dict['SIMULATION']['EIGENSTATES']['EIGENSTATE']['SCALAR_AVERAGE'][0]['MEAN']))
    return Energy

def En1(j, n, m, folder='t-1/'):
    Energy = [] 
    path = "C:/Users/Mindelly/Documents/GitHub/diagrams2/w2/" + folder
    for i in range(n, m, 4):
        file_name = str(i) +'-'+ str(j) + ".task1.out.xml"
        with open(path + file_name) as xml_file:
            data_dict = xmltodict.parse(xml_file.read())
        Energy.append(float(data_dict['SIMULATION']['EIGENSTATES']['EIGENSTATE']['SCALAR_AVERAGE'][0]['MEAN']))
        j = j + 4
    return Energy


def get_result_from_txt(file):
    in_file = open(file, 'r')
    lines = []
    for line in in_file:
        lines.append(line.split())

    for x in range(len(lines)):
        lines[x][0] = int(lines[x][0])
        lines[x][1], lines[x][2] = float(lines[x][1]), float(lines[x][2])

    b = sorted(lines, key=lambda s: s[0])


    mu = []
    h = []
    for x in range(len(b)):
        mu.append(b[x][1])
        h.append(b[x][2])
    return (mu, h)

def mu2(h):
    return - h + e10

def mu3(h):
    return -h - e10
```

```python
color1 = 'limegreen'
color3 = 'tomato'
color2 = 'black'
```

```python
mu, h = get_result_from_txt('result-W=2-0.5.txt')
mu1, h1 = get_result_from_txt('result-ED-PP-W=2-t1.txt')

e11 = En(1, 1, 2, 't-0.5/')[0]
e10 = En(0, 1, 2, 't-0.5/')[0]
h2 = np.arange(h[0], 2.7, 0.1)
h3 = np.arange(5.3, 8, 0.2)

plt.figure(figsize=(14, 9))
xscale = 7.6

plt.plot(h, mu, 'o--', color=color1, label='${t_\perp=0.5}$')
plt.plot(mu1, h1, 'o--', color=color1)
plt.plot(h2, mu2(h2), '-', color=color1)
plt.plot(h3, mu3(h3), '-', color=color1)
plt.axhline(y=(e11/2), color=color1, xmax=(h[0]/xscale))

plt.xlabel('$h$', size=20)
plt.ylabel('$\mu$', size=20)
plt.axis([0, xscale, -4.5, -3.5])


mu, h = get_result_from_txt('result-W=2-0.2.txt')
mu1, h1 = get_result_from_txt('result-ED-PP-02.txt')
e11 = En(1, 1, 2, 't-0.2/')[0]
e10 = En(0, 1, 2, 't-0.2/')[0]

h2 = np.arange(h[0], 2.7, 0.1)
h3 = np.arange(5.3, 8, 0.2)

plt.plot(h, mu, 'o--', color=color2, label='${t_\perp=0.2}$')
plt.plot(mu1, h1, 'o--', color=color2)
plt.plot(h2, mu2(h2), '-', color=color2)
plt.plot(h3, mu3(h3), '-', color=color2)
plt.axhline(y=(e11/2), color=color2, xmax=(h[0]/xscale))
plt.xlabel('$h$', size=20)
plt.ylabel('$\mu$', size=20)
plt.axis([0, xscale, -4.5, -3.5])


e1 = En(0, 2, 79)
e2 = En(1, 3, 80)
e3 = En(1, 1, 78)
e4 = En1(1, 1, 38)
e5 = En1(2, 2, 39)
e6 = En1(0, 2, 39)
e4.append(En(39, 39, 40)[0])
e5.append(En(40, 40, 41)[0])
e6.append(En(38, 40, 41)[0])

h = []
mu = []
for x in range(len(e1)):
    h.append((e3[x] - e1[x]) / -2)
    mu.append((e2[x] - e1[x]) / 2)

h1 = []
mu1 = []
for x in range(len(e4)):
    h1.append((e6[x] - e4[x]) / 2)
    mu1.append((e5[x] - e4[x]) / 2)
    
e10 = En(0, 1, 2)[0]
h2 = np.arange(h[0], 2.7, 0.1)
h3 = np.arange(5.3, 8, 0.2)
plt.plot(h, mu, 'o--', label='${t_\perp=1}$', color=color3)
plt.plot(h1, mu1, 'o--', color=color3)
plt.plot(h2, mu2(h2), '-', color=color3)
plt.plot(h3, mu3(h3), '-', color=color3)
plt.axhline(y=(e4[0]/2), color=color3, xmax=(h[0]/xscale))


plt.tick_params(axis='both', which='major', labelsize=18)
plt.legend(fontsize=16)
plt.savefig('tperp.png')
```
