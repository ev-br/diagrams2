The "polaron" state has the energy $E(N, 1)$. Call the binding energy,$\epsilon_p$, the difference between the energy of the polaron state and the energy of the same state at $U=0$ --- which equals $E(N, 0) + E(0, 1)$. This way,

$$
\epsilon_p := E(N, 1) - E(N, 0) - E(0, 1)
$$
