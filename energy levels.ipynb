```python
import numpy as np
import matplotlib.pyplot as plt
import xmltodict
```

```python
def e(p, t, tperp = 1):
    return -2*t*(np.cos(p)) - tperp

def e1(p, t, tperp = 1):
    return -2*t*(np.cos(p)) + tperp

def Ef(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) - tperp

def Ef1(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) + tperp
```

```python
L = 40
t = 1
tperp = 1
p = np.arange(-10, 10, 0.01)

n = [0]
for b in range(40):
    h = b + 1
    n.append(h)
    n.append(- h)
```

```python
plt.figure(figsize=(15, 17))
plt.plot(p, e(p, t, tperp), '-', color='blue')
plt.plot(p, e1(p, t, tperp), '-', color='lightblue')
plt.axis([-np.pi, np.pi, -3.1, 3])


all_energy = []
for x in range(0, 40):
    N = n[x]
    all_energy.append(Ef1(N, L, t, tperp))
    all_energy.append(Ef(N, L, t, tperp))
all_energy.sort()

i = 0

for each in all_energy:
    if i==16 or i==15:
        c = 'yellow'
    elif i==17 or i==18:
        c = 'orange'
    elif i==19 or i==20 or i==21:
        c = 'red'
    else:
        c = 'lightgreen'
    i = i + 1
    plt.axhline(y = each, color=c)

        
plt.show()

#E = []
#for x in range(40):
##    N = n[x]
#    print(x+1, Ef1(N, L, t, tperp))
#    E.append(Ef1(N, L, tperp))
    

#for x in range(40):
#    N = n[x]
#    print(x+1, Ef(N, L, t, tperp))
#plt.show()


```
