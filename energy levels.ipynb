```python
import numpy as np
import matplotlib.pyplot as plt
import xmltodict
```

```python
#косинус 1 зона
def e(p, t, tperp = 1):
    return -2*t*(np.cos(p)) - tperp
#косинус 2 зона
def e1(p, t, tperp = 1):
    return -2*t*(np.cos(p)) + tperp

def Ef(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) - tperp

def Ef1(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) + tperp
```

```python
L = 40
t = 1
tperp = 1
p = np.arange(-10, 10, 0.01)

n = [0]
for b in range(40):
    h = b + 1
    n.append(h)
    n.append(- h)
```

```python
plt.figure(figsize=(15, 15))
plt.plot(p, e(p, t, tperp), '-', color='blue')
plt.plot(p, e1(p, t, tperp), '-', color='skyblue')
plt.axis([-np.pi, np.pi, -3.2, 3.2])


all_energy = []
for x in range(0, 40):
    N = n[x]
    all_energy.append(Ef1(N, L, t, tperp))
    all_energy.append(Ef(N, L, t, tperp))
all_energy.sort()

i = 0

for each in all_energy:
    if 19 <= i < 61:
        c = 'salmon'
    else:
        c = 'lightgreen'
    i = i + 1
    
    
    plt.axhline(y = each, color=c, label='E')

    
plt.text(0.91, 0.139, '$\epsilon_1$', fontsize=23, transform=plt.gcf().transFigure)
plt.text(0.91, 0.215, '$\epsilon_{10}, \epsilon_{11} $', fontsize=22, transform=plt.gcf().transFigure)
plt.text(0.91, 0.31, '$\epsilon_{16}, \epsilon_{17} $', fontsize=22, transform=plt.gcf().transFigure)
plt.text(0.91, 0.38, '$\epsilon_{20}, \epsilon_{21}, \epsilon_{22} $', fontsize=23, transform=plt.gcf().transFigure)
plt.text(0.91, 0.487, '$\epsilon_{39}, \epsilon_{40} $', fontsize=22, transform=plt.gcf().transFigure)
plt.text(0.91, 0.62, '$\epsilon_{59}, \epsilon_{60} $', fontsize=22, transform=plt.gcf().transFigure)
plt.text(0.91, 0.86, '$\epsilon_{79}, \epsilon_{80} $', fontsize=22, transform=plt.gcf().transFigure)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.xlabel('$k$', size=30)
plt.ylabel('$\epsilon$', size=30)  
plt.savefig('levels-t=1.png')
plt.show()

print(all_energy[18:24])



```

```python
def cos_1(p, t):
    return -2*t*(np.cos(p))
def cos_2(p, t, tperp=1):
    return -2*t*(np.cos(p)) + tperp * np.sqrt(2)
def cos_3(p, t, tperp=1):
    return -2*t*(np.cos(p)) - tperp * np.sqrt(2)

def E(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L))

def E1(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) - tperp * np.sqrt(2)

def E2(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) + tperp * np.sqrt(2)


```

```python
L = 40
t = 1
tperp = 1
p = np.arange(-10, 10, 0.01)

n = [0]
for b in range(40):
    h = b + 1
    n.append(h)
    n.append(- h)
```

```python
all_energy = []
for x in range(0, 40):
    N = n[x]
    all_energy.append(E1(N, L, t))
    all_energy.append(E(N, L, t))
    all_energy.append(E2(N, L, t))
all_energy.sort()
```

```python
plt.figure(figsize=(17, 14))
plt.plot(p, cos_1(p, t), '-', color='dodgerblue')
plt.plot(p, cos_2(p, t), '-', color='blue')
plt.plot(p, cos_3(p, t), '-', color='skyblue')
plt.axis([-np.pi, np.pi, -4, 4])
i = 0
for each in all_energy:
    if 17 <= i < 42:
        c = 'red'
    elif 42 <= i < 78:
        c = 'orange'
    elif 78 <= i < 103:
        c = 'yellow'
    else:
        c = 'lightgreen'
    i += 1
    plt.axhline(y = each, color=c)
    
plt.text(0.91, 0.169, '$\epsilon_1$', fontsize=23, transform=plt.gcf().transFigure)    
plt.text(0.91, 0.31, '$\epsilon_{17} $', fontsize=22, transform=plt.gcf().transFigure)
plt.text(0.91, 0.44, '$\epsilon_{42} $', fontsize=22, transform=plt.gcf().transFigure)
plt.text(0.91, 0.55, '$\epsilon_{78} $', fontsize=22, transform=plt.gcf().transFigure)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.xlabel('$k$', size=30)
plt.ylabel('$\epsilon$', size=30)  
plt.savefig('levels-w=3.png')
```

```python
def tob_cos_1(p, t, tperp=1):
    return -2*t*(np.cos(p)) + tperp # вырожденная зона
def tob_cos_2(p, t, tperp=1):
    return -2*t*(np.cos(p)) - 2*tperp


def E1_tob(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) - 2*tperp

def E2_tob(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) + tperp
def E3_tob(N, L, t, tperp = 1):
    return -2*t*(np.cos(2 * N * np.pi/L)) + tperp
```

```python
L = 40
t = 1
tperp = 1
p = np.arange(-10, 10, 0.01)

n = [0]
for b in range(40):
    h = b + 1
    n.append(h)
    n.append(- h)
```

```python
all_energy_tob = []
for x in range(0, 40):
    N = n[x]
    all_energy_tob.append(E1_tob(N, L, t))
    all_energy_tob.append(E2_tob(N, L, t))
    all_energy_tob.append(E3_tob(N, L, t))

all_energy_tob.sort()
```

```python
plt.figure(figsize=(17, 14))
plt.plot(p, tob_cos_1(p, t), '-', color='dodgerblue')
plt.plot(p, tob_cos_2(p, t), '-', color='blue')
plt.axis([-np.pi, np.pi, -4.3, 3.5])
i = 0
for each in all_energy_tob:
    if 27 <= i < 66:
        c = 'red'
    elif i >= 65:
        c = 'orange'
    #elif 78 <= i < 103:
    #    c = 'yellow'
    else:
        c = 'lightgreen'
    i += 1
    plt.axhline(y = each, color=c)
    
plt.text(0.91, 0.149, '$\epsilon_1$', fontsize=23, transform=plt.gcf().transFigure)    
plt.text(0.91, 0.44, '$\epsilon_{27} $', fontsize=22, transform=plt.gcf().transFigure)
plt.text(0.91, 0.54, '$\epsilon_{65} $', fontsize=22, transform=plt.gcf().transFigure)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.xlabel('$k$', size=30)
plt.ylabel('$\epsilon$', size=30)  
plt.savefig('levels-toblerone.png')
```
