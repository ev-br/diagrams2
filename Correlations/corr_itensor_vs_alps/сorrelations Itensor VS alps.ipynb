# Сравнение $c^{\dagger}_{\uparrow} c_{\uparrow}$, itensor vs alps
## W = 1, L = 16, N_up = 5, N_down = 1

### Корреляции на 1 узле с остальными

```python
file_name1 = 'W1_ver1_corr1.txt'
file_name2 = 'W1_ver2_corr1.txt'
file_name = "W1_L16_5_1_cor1.task1.out.xml"
file1 = open(file_name1).read().splitlines()
file2 = open(file_name2).read().splitlines()
```

```python
# Energy itensor
file1[0]
```

```python
# itensor ver 1
c1 = []
n1 = []
for i in file1[1:16]:
    a = float(i.split('=')[1].split(',')[0][1:])
    b = int(i.split('=')[0].split('(')[2][:-2]) - 1
    c1.append(a)
    n1.append(b)
# itensor ver 2
c2 = []
n2 = []
for i in file2[:16]:
    a = float(i.split(' ')[2].split(',')[0][1:])
    b = int(i.split(' ')[1][2:])
    c2.append(a)
    n2.append(b)
# ALPS
c3 = []
n3 = []
with open(file_name) as xml_file:
    data_dict = xmltodict.parse(xml_file.read())
    for i in range(16):
        a = data_dict['SIMULATION']['EIGENSTATES']['EIGENSTATE']['VECTOR_AVERAGE']['SCALAR_AVERAGE'][i]['MEAN']
        b = data_dict['SIMULATION']['EIGENSTATES']['EIGENSTATE']\
            ['VECTOR_AVERAGE']['SCALAR_AVERAGE'][i]['@indexvalue'][13:-4]
        c3.append(float(a))
        n3.append(int(b))
```

```python
# Energy alps
data_dict['SIMULATION']['EIGENSTATES']['EIGENSTATE']['SCALAR_AVERAGE'][0]['MEAN']
```

```python
file1[1:16]
```

```python
file2[:16]
```

```python
print(data_dict['SIMULATION']['EIGENSTATES']['EIGENSTATE']['VECTOR_AVERAGE']['SCALAR_AVERAGE'][0]['MEAN'])
print(data_dict['SIMULATION']['EIGENSTATES']['EIGENSTATE']['VECTOR_AVERAGE']['SCALAR_AVERAGE'][0]['@indexvalue'][13:-4])
```

```python
%matplotlib notebook
kwds = dict(ls='-', marker='o', lw=2, markersize=4)
kwds1 = dict(ls='--', marker='o', lw=2, markersize=4)
kwds2 = dict(ls='-.', marker='o', lw=2, markersize=4)
plt.plot(n1, c1, **kwds, label='itensor ver1')
plt.plot(n2, c2, **kwds1, label='itensor ver2')
plt.plot(n3, c3, **kwds2, label='ALPS')
plt.xlabel('$i$', size=14)
plt.ylabel('$c^{\\dag}_{\\uparrow} * c_{\\uparrow}$', size=14)
plt.legend()
```
