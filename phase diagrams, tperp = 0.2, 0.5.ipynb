```python
import xmltodict
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_result(file):
    in_file = open(file, 'r')
    lines = []
    for line in in_file:
        lines.append(line.split())

    for x in range(len(lines)):
        lines[x][0] = int(lines[x][0])
        lines[x][1], lines[x][2] = float(lines[x][1]), float(lines[x][2])

    b = sorted(lines, key=lambda s: s[0])


    mu = []
    h = []
    for x in range(len(b)):
        mu.append(b[x][1])
        h.append(b[x][2])
    return (mu, h)

def En(j, n, m):
    Energy = [] 
    path = "C:/Users/Mindelly/Documents/GitHub/diagrams2/w2/t-0.5/"
    for i in range(n, m, 2):
        file_name = str(i) +'-'+ str(j) + ".task1.out.xml"
        with open(path + file_name) as xml_file:
            data_dict = xmltodict.parse(xml_file.read())
        Energy.append(float(data_dict['SIMULATION']['EIGENSTATES']['EIGENSTATE']['SCALAR_AVERAGE'][0]['MEAN']))
    return Energy
```

```python
mu, h = get_result('result-W=2-0.5.txt')
mu1, h1 = get_result('result-ED-PP-W=2-t1.txt')
e11 = En(1, 1, 2)[0]
e10 = En(0, 1, 2)[0]

h2 = np.arange(h[0], 2.7, 0.1)
def mu2(h):
    return - h + e10

h3 = np.arange(5.3, 8, 0.2)
def mu3(h):
    return -h - e10


plt.figure(figsize=(14, 9))
xscale = 7.6

plt.plot(h, mu, 'o--')
plt.plot(mu1, h1, 'o--')
plt.plot(h2, mu2(h2), '-')
plt.plot(h3, mu3(h3), '-')
plt.axhline(y=(e11/2), color='purple', xmax=(h[0]/xscale))

plt.xlabel('$h$', size=20)
plt.ylabel('$\mu$', size=20)
plt.axis([0, xscale, -4.5, -3.5])
plt.title("t_perp = 0.5")
plt.show()
```
