```python
import pyalps
import numpy as np
import matplotlib.pyplot as plt
import pyalps.plot
```

```python
parms = []
for n in [1, 2, 5, 10, 15, 16, 17, 19, 20, 21, 22, 25, 30, 25, 39, 45, 50, 55, 59, 59, 61, 67, 78, 80]:
    parms.append({ 
            'LATTICE'                   : "open square lattice", 
            'MODEL'                     : "fermion Hubbard",
            'CONSERVED_QUANTUMNUMBERS'  : 'Nup, Ndown',
            'Nup_total'                 : n,
            'Ndown_total'               : 0,
            't'                         : 1,
            'U'                         : -7,
            'SWEEPS'                    : 4,
            'NUMBER_EIGENVALUES'        : 1,
            'L'                         : 40,
            'W'                         : 2,
            'MAXSTATES'                 : 400,
            'MEASURE_LOCAL[Local density]'   : 'Nup'})
```

```python
#data = pyalps.loadEigenstateMeasurements(pyalps.getResultFiles(prefix='diffrnt_n')) #старые 
data = pyalps.loadEigenstateMeasurements(pyalps.getResultFiles(prefix='ndiff'))
input_file = pyalps.writeInputFiles('ndiff', parms)
res = pyalps.runApplication('dmrg',input_file,writexml=True)
```
